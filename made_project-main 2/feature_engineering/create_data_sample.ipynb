```python
import random
import pandas as pd
```

#### Generating data

```python
N_GROUPS = 15
GROUP_LINKS_MAX = 10
VISITS_MAX = 5
USERS_NUM = 1000

group_links_num_dict = {}
for group in range(N_GROUPS):
    group_links_num_dict[group] = random.randint(1, GROUP_LINKS_MAX)
```

```python
def gen_behaviour(year, month, day):
    results = ''
    # 14 = two weeks
    for d in range(day - 14, day):
        was_on_site = random.randint(0, 1)
        if was_on_site:
            visits = random.randint(1, VISITS_MAX)
            if results:
                results = f'{results}, [{year}/{month}/{d}, {visits}]'
            else:
                results = f'[{year}/{month}/{d}, {visits}]'
    return f'[{results}]'
```

example of work:

```python
gen_behaviour(2022, 10, 16)
```

```python
with open('sample_data.csv', 'w') as fw:
    fw.write('user_id;target;retrodate')
    for group in range(N_GROUPS):
        for link in range(group_links_num_dict[group]):
            fw.write(f';g{group + 1}(l{group + 1}.{link + 1})')
    fw.write('\n')

    for i in range(USERS_NUM):
        user_id = i
        target = random.randint(0, 1)
        retrodate_y = random.randint(2020, 2021)
        retrodate_m = random.randint(1, 12)
        retrodate_d = random.randint(15, 28)

        retrodate = f'{retrodate_y}/{retrodate_m}/{retrodate_d}'

        string = f'{user_id};{target};{retrodate}'
        for group in range(N_GROUPS):
            for link in range(group_links_num_dict[group]):
                behaviour = gen_behaviour(retrodate_y, retrodate_m, retrodate_d)
                string = f'{string};{behaviour}'
        
        fw.write(string + '\n')
    
    
```
